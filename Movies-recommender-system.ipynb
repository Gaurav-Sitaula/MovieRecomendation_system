```python
# since above list is in the form of string we have to convert it into list, for that we will use
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
#so by importing ast we can convert it into list
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return l
```

```python
#applying the function into the column genre
movies["genres"].apply(convert)
```

```python
movies["genres"]=movies["genres"].apply(convert)
```

```python
def convert1(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            l.append(i["name"])
            counter+=1
        else:
            break
    return l
```

```python
movies["cast"]=movies["cast"].apply(convert1)
```

```python
movies.head(1)
```

```python
movies["crew"][0]
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i["job"]=="Director":
            l.append(i["name"])
            break
    return l
```

```python
movies["crew"]=movies["crew"].apply(fetch_director)
```

```python
print(type(movies["overview"]))
```

```python
movies["overview"]=movies["overview"].astype(str)
unique_types = movies["overview"].apply(lambda x: type(x)).unique()

# Print the unique data types
for data_type in unique_types:
    print(data_type)
```

```python
#def split_string(x):
    #return x.split()
movies["overview"]=movies["overview"].apply(lambda x:x.split())
```

```python
movies.head(2)
```

```python
#Replacing spaces with no sapce.. for eg: in genre science fiction becomes sciencefiction
movies["genres"]=movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies["cast"]=movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"]=movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"]=movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(2)
```

```python
#making new column
movies["tags"]=movies["overview"]+movies["genres"]+movies["cast"]+movies["keywords"]+movies["crew"]
```

```python
#converting tags whch is in list into string
new_movies_df["tags"]=new_movies_df["tags"].apply(lambda x:" ".join(x))
#The string join() method returns a string by joining all the elements of an iterable (list, string, tuple), separated by the given separator
```

```python
#It is recommended to tranform all tags into lowercase
new_movies_df["tags"]=new_movies_df["tags"].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
#stemmig example
#["loving","loves","loved"]
#["love","love","love"]
```

```python
def stem(text):
    y=[]
    for i in text.split(): #text.split converts string to list
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
#Now finding the similarities on the basis of text in the tag
#how can we find?
 # Similarity depends upon the similarity score of the text
    #1 technique:Calculatin number of similar words(Not good technique)
    #2 technique: converting tags into vectors(called as text vectoriation)
        # bag of words, tfidf,word2vec are the methods used in text vectorization
        # During vectorization we do not consider stopwords for eg: and, are, from,to etc so we remove them
    
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
#countvectorizer give a saparse matrix
#now we use cv object
cv.fit_transform(new_movies_df["tags"]).toarray()
cv.fit_transform(new_movies_df["tags"]).toarray().shape
```

```python
vectors=cv.fit_transform(new_movies_df["tags"]).toarray()
```

```python
def recommend(movie):
    movie_index=new_movies_df[new_movies_df["title"]==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_movies_df.iloc[i[0]].title)
```
